<h1>2. What are movies on Netflix having imdb_score>=8?

```python
top10moviesi9 = df[df['imdb_score']>=8].sort_values(by='imdb_score', ascending=False)
top10moviesi9
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'imdb_score', y = 'title', data = top10moviesi9, palette='viridis')
plt.title('Top movies on Netflix having imdb_score>=8')
plt.xlabel('IMDB Score')
plt.ylabel('Movies Names')
# plt.tight_layout()
plt.show()
```

<h1>3. List 10 Drama movies released on Netflix after year 2018 having imdb_score>=7?

```python
top10dramai8 = df[(df['imdb_score']>=7) & (df['genre']=='Drama') & (df['year']>=2018)].sort_values(by='imdb_score', ascending=False).head(10)
top10dramai8
```

```python
top10dramai8.plot(kind='bar', x='title', y='imdb_score', figsize=(10,6))
```

```python
sns.barplot(x = 'imdb_score', y = 'title', data = top10dramai8, palette='viridis')
plt.title('Top 10 Drama TV shows on Netflix having imdb_score>=8 after year 2018')
plt.xlabel('IMDB Score')
plt.ylabel('TV Shows Names')
plt.show()
```

<h1> 4. In which year maximum number of videos released on netflix?

```python
df['year'].value_counts().idxmax()
```

```python
maxNoOfVideos = df['year'].value_counts()
maxNoOfVideos
```

```python
sns.barplot(x = maxNoOfVideos.index, y = maxNoOfVideos.values, palette='viridis')
plt.title('Maximum number of videos released on Netflix in which year?')
plt.xlabel('Year')
plt.ylabel('Number of Videos')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
clr = sns.color_palette('rainbow')

# Format to display count and percentage
autopct = lambda pct: f'{pct:.1f}%\n({round(pct/100.*sum(maxNoOfVideos))})'

plt.pie(maxNoOfVideos, labels=maxNoOfVideos.index, colors=clr, autopct=autopct, startangle=70)
plt.title('Maximum number of videos released on Netflix in which year?')
plt.legend(loc='upper right')
plt.show()
```

<h1> 5. In which genre most of the videos released on netflix after year 2016?

```python
df.groupby('genre')['year'].value_counts()
```

```python
# Filter the DataFrame to include only rows where the year is greater than 2016
filtered_df = df[df['year'] > 2016]

# Group by genre and count the number of occurrences
result = filtered_df.groupby('genre')['year'].count()
result
```

```python
result.idxmax()
```

```python
# Filter the DataFrame to include only rows where the year is greater than 2016
filtered_df = df[df['year'] > 2016]

# Group by genre and count the number of occurrences
result = filtered_df.groupby('genre')['year'].value_counts()
result
```

```python
result.unstack().plot(kind='bar', stacked=True, figsize=(20, 10))

plt.title('In which genre most of the videos released on netflix after year 2016?')
plt.xlabel('Genre')
plt.ylabel('Number of Videos')
plt.show()
```
